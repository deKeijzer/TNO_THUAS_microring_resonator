# Load .mat file in Python

```python
def load_wavelength_shift_file(location,file):
    mat = loadmat(location+file)
    
    # select the wanted dict and convert dtypes of structures to dicts
    mdata = mat['ISW_result']  # variable in mat file
    mdtype = mdata.dtype  # dtypes of structures are "unsized objects"
    ndata = {n: mdata[n][0, 0] for n in mdtype.names}

    # finally select the beloved data
    time = ndata["time"]
    sensor = ndata["sensor"] # wavelength shift

    x = time.flatten()
    y = sensor[:,0].flatten()
    
    print(x.shape)
    
    df = pd.DataFrame([x,y]).T
    df.columns = ["x", "y"]
    
    return df
```

# Complete plot

```python
file = "chip14_P4_poort5 water_20191211_beker1_4_result.mat"
df = load_wavelength_shift_file(location,file)
```

```python
plt.figure(figsize=(10,5))

plt.plot(df.x,df.y, '.-', color='red', ms=2, linewidth=1, label="Measurement sequence 4")

plt.xlabel("Time $t$ $\quad$ [s]")
plt.ylabel("Wavelength shift $\quad$ [pm]")
plt.legend(loc=(0,1))

plt.grid()

#plt.savefig("C:\\Users\\BDK\\Desktop\\Nanotechnologie\\FRESCO\\figuren\\beker1_4_volledig.png", dpi=300)
```
